# Tạo kiểu là gì và tại sao quan tâm?
Ý tưởng căn bản đằng sau việc tạo kiểu là một người dùng muốn thay đổi cách dữ liệu được biểu diễn nhưng vẫn giữ lại định dạng phía dưới cho việc thay đổi sau này.

Ví dụ tạo kiểu đơn giản nhất là sử dụng một biểu tượng tiền tệ khi làm việc với các giá trị tiền tệ. Ví dụ, nếu dữ liệu của bạn chứa giá trị 24.00, bạn không muoons biết ngay lập tức nếu giá trị là ở dạng dollars, pounds, euros hay đơn vị tiền tệ khác. Nếu số đó là $25 khi đó ý nghĩa nó rõ ràng. 

Phần trăm là một ví dụ hữu dụng khác khi định dạng output làm cho nó dễ hiểu hơn phân tích phía dưới. Ví dụ, cái nào sau đây làm ta hiểu nhanh hơn: .0.5 hay 5\% Sử dụng kí hiệu phần trăm làm cho ta dễ phân tích dữ liệu hơn. 

Điều cần nhớ ở đây là làm kiểu biểu diễn dữ liệu vì vậy một con người có thể đọc nhưng nó vẫn giữ cho dữ liệu cùng một kiểu dữ liệu pandas giống như vậy vì vậy bạn có thể thực hiện các phép toán, ngày tháng, chuỗi kí tự trên nó.

Tạo kiểu pandas còn cung cấp thêm nhiều công cụ nâng cao để thêm màu hoặc các thành phần hình hoạ khác cho output. 

# Làm kiểu dữ liệu
Hãy bắt đầu bằng việc nhìn vào dữ liệu. Trong ví dụ này chúng ta sẽ sử dụng dữ liệu sales 2018 cho một tổ chức tưởng tượng nào đó. Chúng ta sẽ giả bộ như là một người phân tích tìm kiếm cho các xu thế sales cấp cao cho năm 2018. 

```python
import numpy as np
import pandas as pd

df = pd.read_excel("../pbpython/data/2018_Sales_Total.xlsx")
df.head()
```

Đối với dữ liệu trên, bạn có thể làm một tổng kết nhanh bao nhiêu khách hàng đã mua từ chúng ta và trung bình lượng mua.

```python
 df.groupby("name")["ext price"].agg(["mean", "sum"])
```

Khi nhìn vào dữ liệ như trên, rất khó có thể hiểu được mức độ của các số bởi vì bạn có các số có 6 chữ số. Còn nữa, ta không biết được số tiền kia là ở đơn vị tiền tệ nào.

```python
(df.groupby("name")["ext price"].agg(["mean", "sum"]).style.format("${0:,.2f}"))
```

Sử dụng hàm **format**, bạn có thể sử dụng tất cả các tính năng của các công cụ formating chuỗi văn bản trên dữ liệu này. Trong trường hợp này, chúng ta sử dụng **${0:,.2f}** để đặt dấu dollar vào đầu, thêm dấu hỏi và làm tròn kết quả tới 2 con số thập phân. 

Nếu bạn muốn nhìn vào tổng số lượng sales mỗi tháng, bạn có thể sử dụng **grouper** để tổng kết theo tháng và tính toán mỗi tháng như là phần trăm của tổng lượng sales của năm

```python
monthly_sales = df.groupby([pd.Grouper(key='date', freq='M')])['ext price'].agg(['sum']).reset_index()
monthly_sales['pct_of_total'] = monthly_sales['sum'] / df['ext price'].sum()
monthly_sales
```

Ta biết cáhc để tạo kiểu các số nhưng bây giờ ta có một sự kết hợp ngày tháng, phần trăm và tiền tệ. May mắn là chúng ta có thể sử dụng một tự điển để định nghĩa một formatting string duy nhât cho mỗi column. Điều này sẽ giúp ích được ta rất nhiều.

```python
format_dict = {'sum':'${0:,.0f}', 'date': '{:%Y}', 'pct_of_total': '{:.2%}'}
monthly_sales.style.format(format_dict).hide_index()
```

Khi ta phát triển report cuối cùng, có một kiểu format như vậy sẽ rất hữu dụng. Trong ví dụ trên ta khoong thấy các index vì đã sử dụng hàm **hide_index**.

Ngoài việc tạo kiểu cho số, ta có thể tạo kiểu cho các cells trong DataFrame. Nào hãy cùng highlght số lớn nhất màu xanh và số nhỏ nhất màu Trinidad.

```python
(monthly_sales
 .style
 .format(format_dict)
 .hide_index()
 .highlight_max(color='lightgreen')
 .highlight_min(color='#cd4f39'))
```

Một hàm hữu dụng khác là **background_gradient** dùng để highlight dải dải trị trong một cột.

```python
(monthly_sales.style
 .format(format_dict)
 .background_gradient(subset=['pct_of_total'], cmap='BuGn'))
```

Ví dụ phía trên cho ta biết được cách sử dụng của parameter **subset** để áp dụng các hàm tới chỉ một cột đơn lẻ của dữ liệu. Ngoài ra, tham số **cmap** cho phép chúng ta chọn một màu bảng màu cho khuynh độ.

# Tạo kiểu với Bars
Hàm tạo kiểu còn hỗ trợ việc vẽ các bar charts bên trong mỗi cột

```python
(monthly_sales
 .style
 .format(format_dict)
 .hide_index()
 .bar(color='#FFA07A', vmin=100_000, subset=['sum'], align='zero')
 .bar(color='lightgreen', vmin=0, subset=['pct_of_total'], align='zero')
 .set_caption('2018 Sales Performance'))
```

Trong ví dụ trên ta có thể dùng cả đặt caption cho hình sử dụng hàm **set_caption**
